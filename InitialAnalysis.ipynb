```python
# notebook to produce initial correlations and visualizations

import sklearn
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing as prep
import DataCleaning
import numpy as np
```

```python
allBitcoinData = DataCleaning.getBitcoinData()
allEthereumData = DataCleaning.getEthereumData()
allDogeData = DataCleaning.getDogeData()
allShibaData = DataCleaning.getShibaData()
```

```python
import matplotlib.dates as mdates
fig, ax = plt.subplots(2, 2, figsize=(15,15))
plt.style.use('dark_background')

xfmt = mdates.DateFormatter('%Y-%b')
months = mdates.MonthLocator()
ax[0, 0].plot(allEthereumData.index, allEthereumData["standardized_price"], label="Price")
ax[0, 1].plot(allEthereumData.index, allEthereumData["standardized_price"], label="Price")
ax[1, 0].plot(allEthereumData.index, allEthereumData["standardized_price"], label="Price")
ax[1, 1].plot(allEthereumData.index, allEthereumData["standardized_price"], label="Price")
ax[0, 0].plot(allEthereumData.index, allEthereumData["standardized_txs"], label="Transactions")
ax[0, 1].plot(allEthereumData.index, allEthereumData["standardized_posts"], label="Posts")
ax[1, 0].plot(allEthereumData.index, allEthereumData["standardized_tweets"], label="Tweets")
ax[1, 1].plot(allEthereumData.index, allEthereumData["standardized_trends"], label="Trends")
ax[0, 0].set_ylabel("Standardized Value")
ax[1, 0].set_ylabel("Standardized Value")
ax[1, 0].set_xlabel("Date")
ax[1, 1].set_xlabel("Date")
for currAx in ax.flat:
    currAx.xaxis.set_major_locator(months)
    currAx.xaxis.set_major_formatter(xfmt)
    currAx.legend()
    currAx.tick_params(labelrotation=45)
fig.suptitle("Ethereum Correlations", size='xx-large')
```
