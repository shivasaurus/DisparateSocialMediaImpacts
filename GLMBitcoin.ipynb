```python
import DataCleaning
import Transformation
allBitcoinData = DataCleaning.getBitcoinData()
```

```python
import statsmodels.api as sm
import pandas as pd
```

```python
X = sm.add_constant(allBitcoinData[['transactions_float', 'tweet_count', 'trends_float', "posts_float"]])
Y = allBitcoinData["Close"]
lognormal_model = sm.GLM(Y, X, family=sm.families.InverseGaussian())
lognormal_results = lognormal_model.fit()
print(lognormal_results.summary())
```

```python
trX = sm.add_constant(Transformation.repeated_gs_removal(["posts_float", "tweet_count", "trends_float"], ["transactions_float"], allBitcoinData))                                                                                                                                    
                                       
```

```python
idx = pd.date_range('2020-08-02', '2022-02-28')
X = trX
X.index = idx
Y = allBitcoinData["standardized_price"]
X
```

```python
lognormal_model = sm.GLM(Y, X, family=sm.families.Gaussian(sm.families.links.log()))
```

```python
lognormal_results = lognormal_model.fit()
```

```python
print(lognormal_results.summary())
```
